```python
x= df[["score_1","age","rep_year","past_billings_cnt","close_loans_cnt","payment_type_2","payment_type_3"]]
y= df["bad_flag"]

y
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
X_resampled.shape
```

```python
y_resampled.shape
```

```python
from sklearn.model_selection import train_test_split

X_resampled_train, X_resampled_test, y_resampled_train, y_resampled_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42
)
```

```python
X_resampled_train.shape, X_resampled_test.shape, y_resampled_train.shape, y_resampled_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Random Forest
rf_model = RandomForestClassifier(n_estimators=100, max_depth=5)

# Decision Tree (for AdaBoost)
dt_stump = DecisionTreeClassifier(max_depth=1)

# AdaBoost with Decision Tree as base estimator
ada_model = AdaBoostClassifier(estimator=dt_stump, n_estimators=50, learning_rate=0.1)
# Support Vector Machine
svm_model = SVC(kernel='rbf', C=1.0)

# Logistic Regression
logreg_model = LogisticRegression(max_iter=1000)
```

```python
rf_model.fit(X_resampled_train, y_resampled_train)
ada_model.fit(X_resampled_train, y_resampled_train)
svm_model.fit(X_resampled_train, y_resampled_train)
logreg_model.fit(X_resampled_train, y_resampled_train)
```

```python
models = {
    'Random Forest': rf_model,
    'AdaBoost': ada_model,
    'SVM': svm_model,
    'Logistic Regression': logreg_model
}

for name, model in models.items():
    model.fit(X_resampled_train, y_resampled_train)
    score = model.score(X_resampled_test, y_resampled_test)
    print(f'{name} Accuracy: {score:.2f}')
```
